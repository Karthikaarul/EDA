```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 #Update this with the correct path to your file
import pandas as pd
file_path = 'C:\\Users\\DELL\\Desktop\\Global YouTube Statistics.csv'  # Update this with the correct path to your file
data = pd.read_csv(file_path, encoding='latin1')
```

```python
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
stats_summary = data.describe(include='all')
print("Statistical Summary:\n", stats_summary)
```

```python
import seaborn as sns
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(data['subscribers'], bins=30, kde=True)
plt.title('Distribution of Subscribers')
plt.xlabel('Subscribers')
```

```python

plt.subplot(1, 2, 2)
sns.histplot(data['video views'], bins=30, kde=True)
plt.title('Distribution of Video Views')
plt.xlabel('Video Views')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y=data['category'], order=data['category'].value_counts().index)
plt.title('Most Common Categories of YouTube Channels')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y=data['Country'], order=data['Country'].value_counts().index)
plt.title('Distribution of YouTube Channels by Country')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['created_year'].dropna(), bins=30, kde=True)
plt.title('Trends in Creation of YouTube Channels Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
top_10_channels = data.nlargest(10, 'subscribers')
plt.figure(figsize=(14, 6))
sns.barplot(x='subscribers', y='Youtuber', data=top_10_channels)
plt.title('Top 10 YouTube Channels by Subscribers')
plt.xlabel('Subscribers')
plt.ylabel('YouTube Channel')
plt.show()
```

```python
top_10_countries = data['Country'].value_counts().head(10)
plt.figure(figsize=(14, 6))
sns.barplot(x=top_10_countries.values, y=top_10_countries.index)
plt.title('Top 10 Countries by Number of YouTube Channels')
plt.xlabel('Number of YouTube Channels')
plt.ylabel('Country')
plt.show()
```
