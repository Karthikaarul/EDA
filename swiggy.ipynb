```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df=pd.read_csv("C:\\Users\\DELL\\Desktop\\swiggy.csv")
```

```python
summary_stats = df.describe()
print(summary_stats)
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Plot 1: Distribution of Average Ratings
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.histplot(df['Avg ratings'], bins=30, kde=True)
plt.title('Distribution of Average Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Delivery time'], bins=30, kde=True)
plt.title('Distribution of Delivery Times')
plt.xlabel('Delivery Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Price'], bins=30, kde=True)
plt.title('Distribution of Meal Prices')
plt.xlabel('Meal Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='City', data=df, order=df['City'].value_counts().index)
plt.title('Count of Restaurants by City')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Price', y='Avg ratings', data=df)
plt.title('Relationship between Meal Prices and Average Ratings')
plt.xlabel('Meal Price')
plt.ylabel('Average Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Total ratings'], bins=30, kde=True)
plt.title('Distribution of Total Ratings')
plt.xlabel('Total Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
top_10_restaurants = df.nlargest(10, 'Avg ratings')
plt.figure(figsize=(12, 6))
sns.barplot(x='Avg ratings', y='Restaurant', data=top_10_restaurants)
plt.title('Top 10 Restaurants by Average Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Restaurant')
plt.show()
```
